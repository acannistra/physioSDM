```r
library(dplyr)
library(ggplot2)
```

```r
sigmoid.tmax <- function(env, tmax) ifelse(env<tmax-10, 1.0, exp(-(env-tmax+10)/5)) #max  
sigmoid.tmin <- function(env, tmin) ifelse(env<tmin   , 0.1, 1- exp(-(env-(tmin)/9000) ) )
```

```r
temps = seq(-10, 100, 0.5)
```

```r
plot(temps, sigmoid.tmin(temps, 0.5)*sigmoid.tmax(temps, 40), type='o')
```
